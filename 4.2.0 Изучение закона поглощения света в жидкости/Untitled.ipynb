```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data_df = pd.read_excel('Book1.xlsx')
data_df
```

```python
plt.figure(figsize=(20,10))
plt.plot(data_df['h (cm)'], data_df.I)
plt.show()
```

```python
x1 = data_df[data_df['Series'] == 1]['h (cm)'].apply(lambda x: x * 2)
x2 = data_df[data_df['Series'] == 2]['h (cm)'].apply(lambda x: x * 2)
x3 = data_df[data_df['Series'] == 3]['h (cm)'].apply(lambda x: x * 2)

y1 = data_df[data_df['Series'] == 1]['I'].apply(lambda i: math.log(float(i)/255.0))
y2 = data_df[data_df['Series'] == 2]['I'].apply(lambda i: math.log(float(i)/255.0))
y3 = data_df[data_df['Series'] == 3]['I'].apply(lambda i: math.log(float(i)/255.0))

print(x3)
print(y3)
```

```python
with sns.plotting_context("poster"), sns.axes_style("darkgrid"):
    plt.figure(figsize=(15,10))
    
    series_3 = plt.scatter(x3, y3, marker='v')
    series_2 = plt.scatter(x2, y2, marker='o')
    series_1 = plt.scatter(x1, y1, marker='D')
    
    plt.xlabel('$2h$, ')
    plt.ylabel('$ln(I /I_0)$')
    plt.legend([series_1, series_2, series_3], [' 1',' 2',' 3'])

    plt.savefig('scatter.png', dpi=450)
```

```python
with sns.plotting_context("poster"), sns.axes_style("darkgrid"):
    plt.figure(figsize=(15,10))
    
    series_3 = plt.scatter(x3, y3, marker='v')
    series_2 = plt.scatter(x2, y2, marker='o')
    series_1 = plt.scatter(x1, y1, marker='D')
    
    x = np.linspace(1, 15, 100)
    approx_1 = plt.plot(x, -0.113439 * x + 0.474655) #da=0.0416658, db=0.372671
    approx_2 = plt.plot(x, -0.139116 * x + 0.551565) #da=0.0418133, db=0.37399
    approx_3 = plt.plot(x, -0.152563 * x + 0.545556) #da=0.0344131, db=0.3078
    
    plt.xlabel('$2h$, ')
    plt.ylabel('$ln(I /I_0)$')
    plt.legend([series_1, series_2, series_3], ['-0.11x + 0.48','-0.14x + 0.55','-0.15x + 0.54'])
    
    plt.savefig('with_lines.png', dpi=450)
```
