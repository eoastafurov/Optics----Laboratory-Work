```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data_df = pd.DataFrame({
    'm':[-4, -3, -2, -1, 1, 2, 3, 4],
    'dx':[1.3, 0.9, 1.1, 0.8, 0.8, 1.0, 1.1, 1.2]
})
data_df['dx'] = data_df['dx'].apply(lambda x: x * 0.01)
print(data_df.dx)
```

```python
with sns.plotting_context("poster"), sns.axes_style("darkgrid"):
    plt.figure(figsize=(15, 10))
    points = plt.scatter(data_df.m, data_df.dx)
    plt.xlim(-6, 6)
    plt.ylim(-0.01, 0.03)
    
    x = np.linspace(-10, 10, 100)
    line = plt.plot(x, 4.62593e-19 * x + 0.01025) #db=0.0013994
    
    plt.legend([points], ['$0 \cdot  x + 0.01$'])
    
    plt.xlabel('m, ')
    plt.ylabel('$\Delta x$, ')
    plt.title('      m')
    plt.savefig('graph.png', dpi=450)
    
    plt.show()
```
